```python
#ライブラリインポート
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import pulp
from cvxopt import matrix
import cvxopt
```

```python
#データフレーム化
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df
```

```python
# クラスタサイズ設定
K = 50
#クラスタサイズの幅(今回は18と22)
alpha = 0.3
alpha_d = 0.3
```

```python
# サンプル値（重心の初期値）を取るためのリストを作成
list_sample = df.to_numpy().tolist()

# データフレームをnp.array型変換
X = np.array(df)

# Xのサイズと特徴量の数を取得
X_size, n_features = X.shape
X.shape

#クラスタ数
c = int(X_size / K)

#目的関数の最小値
min_J_all=0

#最初の時のクラスタ
min_cluster=[0 for i in range(X_size)]
```

```python
#初期値を変えて複数回
for f in range(100):
    
    v_first = []
    # 入力値の中から重複の内容にサンプリング
    v_first = random.sample(list_sample, c)
    # ↑ここまでで重心の初期値を定義
    #許容範囲のκを設定
    #誤差の初期値を設定
    kappa = []
    epsilon = []

    min_J = 0
    min_u = np.zeros(X_size)
    for i in range(X_size):
    
        ep = []
        ka = []
        for j in range(n_features):
            ka.append(random.uniform(0.0001 , 0.0002))
            ep.append(random.uniform(ka[j] * (-1) , ka[j]))
        epsilon.append(ep)
        kappa.append(ka)
#-----------------------------ここまでOK-------------------------------------------   
    # 重心の初期値を代入
    v = v_first
    
    #-------------------ここから繰り返し-------------------------
    for w in range(1000):
        # 各データ点が属するクラスを入れる変数作成
        u = np.zeros(X_size)
        
        #二次計画法で使用するP行列を作成
    
        #pにはd11,d12,d13,,,,d21,d22,,,dn1,,dncの順でリスト作る
        p=[]
        for i in range(X_size):
            for j in range(c):
                p.append(np.sum((v[j]-X[i]-epsilon[i])**2))
        #三項演算子でデータ点をクラスタ中心にしている場合、距離dを少しずらす←クラスタ中心をずらした方がいいかも
        p = [0.0000000001 if i == 0 else i for i in p]
        #Pは1/2倍しておく
        p_1 = 1/2 * np.diag(p)
        P = matrix(p_1.astype('float'))
        #-------------------ここまでP------------------------------------------------------

        #-------------------ここからG------------------------------------------------------
        g = [-1 for i in range(X_size*c)]
        G = matrix(np.diag(g).astype("float"))
        #-------------------ここからh------------------------------------------------------
        h = matrix(np.zeros((X_size,1)).astype('float'))
        
        #bを作るための配列作成
        b_pre = [1 for i in range(X_size)]
        for i in range(c):
            b_pre.append(K)
            
        #q,b作成
        q = matrix(np.zeros(X_size*c).astype('float'))
        b = matrix(np.array(b_pre).astype('float'))
        
        #A作成
        s_pre1=[]
        s_pre=[]
        for i in range(X_size):
            for j in range(X_size*c):
                if j==3*i or j==3*i+1 or j==3*i+2:
                    s_pre1.append(1)
                else:
                    s_pre1.append(0)
            s_pre.append(s_pre1)
            s_pre1=[]

        for i in range(c):
            for j in range(X_size*c):
                if i % 3 == 0 and j % 3 == 0:
                    s_pre1.append(1)
                elif i % 3 == 0 and j % 3 ==1:
                    s_pre1.append(0)
                elif i % 3 == 0 and j % 3 ==2:
                    s_pre1.append(0)
                elif i % 3 == 1 and j % 3 ==0:
                    s_pre1.append(0)
                elif i % 3 == 1 and j % 3 ==1:
                    s_pre1.append(1)
                elif i % 3 == 1 and j % 3 ==2:
                    s_pre1.append(0)
                elif i % 3 == 2 and j % 3 ==0:
                    s_pre1.append(0)
                elif i % 3 == 2 and j % 3 ==1:
                    s_pre1.append(0)
                elif i % 3 == 2 and j % 3 ==2:
                    s_pre1.append(1)
            s_pre.append(s_pre1)
            s_pre1=[]
        s = np.array(s_pre).astype('float')
        A = matrix(s)
        #---------------------ここまでA作成----------------------------
        
        #解く
        sol = cvxopt.solvers.qp(P,q,A=A,b=b)
        #連続値になっている帰属度を0、1に変換（一番大きなクラスタに配属）
        u = np.array(sol['x'])

        cluster = [0 for i in range(X_size)]
        u_list = u.tolist()
        for i in range (X_size):
            u_list=[u[i*3],u[1+i*3],u[i*3+2]]
            if  u_list.index(max(u_list))==0:
                cluster[i] = 0
            elif u_list.index(max(u_list))==1:
                cluster[i] = 1
            else:
                cluster[i] = 2
        
        #ここからJ（目的関数の値の計算）
        J=0
        for i in range(X_size):
            for j in range(c):
                J = J + u[3*i+j][0]*u[3*i+j][0]*np.sum((v[j]-X[i]-epsilon[i])**2)
        if J < min_J and w != 0:
            min_J = J
            min_u = u
        elif w == 0:
            min_J = J
            min_u = u
        
#ーーーーーーーーーーーーここから許容範囲（誤差）の更新ーーーーーーーーーーーーーーーーーーーーーー
        for i in range(X_size):
            for j in range(n_features):
        #絶対値のない方
                dis_bunsi = (X[i][j]-v[int(cluster[i])][j])
        #絶対値ある方
                dis_abs = abs(X[i][j]-v[int(cluster[i])][j])
        
        #alphaの部分
                if dis_abs != 0 and kappa[i][j]/dis_abs < 1:
                    alpha = kappa[i][j]/dis_abs
                else:
                    alpha = 1
                epsilon[i][j] = (-1) * alpha * dis_bunsi
#ーーーーーーーーーーーーーーここまでーーーーーーーーーーーーーーーーーーーーーーーーーーーーーーー
#ーーーーーーーーーーここから重心の更新ーーーーーーーーーーーーーーーーーーーーーーーーーーーーーー
        new_v = np.zeros((c, n_features))
        for j in range(c):    
            new_v[j] = X[np.array(cluster) == j].mean(axis=0)
        if (new_v != v).all():
            v=new_v
        else:
            break
#----------------------------------------------------------------------------------
 
#目的関数が最小かどうかの判断
    if min_J < min_J_all and f != 0:
        min_J_all = min_J
        min_cluster = cluster
        print("目的関数："+str(min_J_all))
    elif f == 0:
        min_J_all = min_J
        min_cluster = cluster
        print("目的関数："+str(min_J_all))

        

P
```
