```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cp
#sysはコマンドライン引数
import sys
import csv
np.seterr(divide='ignore', invalid='ignore')
#csvファイルの読み込み
iris= pd.read_csv("../dataset/iris-dataset.csv")
df = iris.drop('species', axis=1)
#df= pd.read_csv("../dataset/iris-dataset.csv")
#dfをnumpyで扱えるnp.ndarray変換
X = df.values
#配列の行数、列数の格納
X_size,dimension = X.shape
#クラスタ数宣言
c = 3
#Kはクラスタサイズ 
K = X_size /c
max_iter = 1
best_J = 10000
from sympy import diff, symbols, solve
```

```python
X.shape
special_X = X[np.random.choice(X_size,50)]
special_X.shape #データサイズを50,4まで小さくしておく
X_size,dimension = special_X.shape
```

ラグランジュのい
from sympy import diff, symbols, solve

# 変数を定義
x, y, z, l = symbols("x y z l")

# 関数ｆを定義
f = x * y**2 * z**3

# 制約条件ｇを定義
g = x + y + z - 1

# 定理式に代入
theor = f - l * g

# 各変数で偏微分する
diffx = diff(theor, x)
diffy = diff(theor, y)
diffz = diff(theor, z)
diffl = diff(theor, l)

# 連立方程式を解く
res = solve([diffx, diffy, diffz, diffl])

```python
#変数を定義(u,l,m)
import sympy
chrs = sympy.symbols("u_{1:51\,1:4}")
u_list = list(chrs)
print(u_list)
# [a_{1,1}, a_{1,2}, a_{1,3}, a_{2,1}, ..., a_{3,3}]
#ラムダはlで定義
chrs = sympy.symbols("l_{1:51}")
l_list = list(chrs)
print(l_list)
#μはmで定義
chrs = sympy.symbols("m_{1:3}")
m_list = list(chrs)
print(m_list)
```

```python
len(l_list)
```

```python
centroids = X[np.random.choice(X_size,c)]
#dにはd11,d12,d13,d14,d21,d22,,,dn1,,dncの順でリスト作る
d=[]
for i in range(X_size):
    for j in range(c):
        d.append(np.sum((X[i]-centroids[j])**2))
```

```python
## 
```

```python
#目的関数 Jを定義
J = 0
for i in range(X_size*c):
    J = J + u_list[i]**3 * d[i]
print(J)
```

## 制約条件を作る

sum(i=1,,c)u[k][i]-1=0

```python
seiyaku_1= []
for i in range(0,len(u_list),c):
    seiyaku_1.append(u_list[i]+u_list[i+1]+u_list[i+2]-1)
```

sum(k=1...X_szie)u[k][i]-K=0

```python
s = [0 for i in range(c-1)]#要修正
for i in range(0,len(u_list),c):
    s[0] = s[0] + u_list[i]
    s[1] = s[1] + u_list[i+1]
    #s[2] = s[2] + u_list[i+2]
# マイナスKする関数
def calc_K(n):
    return n-3
# map関数を使用してdata1の要素を全て倍にしてdata2に格納する場合
data2 = map(calc_K,s)
seiyaku_2 = list(data2)   # mapをlistに変換
seiyaku_2
```

## 定理式に代入
theor = J - l * seiyaku_1 - m * seiyaku_2

```python
theor = J
for i in range(len(l_list)):
    theor = theor + l_list[i] * seiyaku_1[i]
for i in range(len(m_list)):
    theor = theor + m_list[i] * seiyaku_2[i]
theor
```

各変数で微分

diffx = diff(theor, x)
diffy = diff(theor, y)
diffz = diff(theor, z)
diffl = diff(theor, l)

```python
answer = []
for i in range(len(u_list)):
    answer.append(diff(theor,u_list[i]))
for i in range(len(l_list)):
    answer.append(diff(theor,l_list[i]))
for i in range(len(m_list)):
    answer.append(diff(theor,m_list[i]))
```

```python
res = solve(answer)
```
