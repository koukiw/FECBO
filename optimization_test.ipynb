scipyのoptimizationを色々試してみた

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(context="paper" , style ="whitegrid",rc={"figure.facecolor":"white"})

from scipy.optimize import minimize
import scipy as sp
import matplotlib.patches as mpatches
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
def f(x,args ):
    v = (args[0] - x[0])**2 + args[1]*(x[1] - x[0]**2)**2
    return(v)

def grad(x,args):
    delx = 2*(x[0]- args[0]) + 4*args[1]*x[0]*(x[0]**2 - x[1])
    dely = 2*args[1]*(x[1] - x[0]**2)
    g = np.array([delx,dely])
    return(g)

def hess(x,args):
    delxx = 2 + 12*args[1]*x[0]**2 - 4*args[1]*x[1]
    delxy = -4*args[1]*x[0]
    delyx = delxy
    delyy = 2*args[1]
    h = np.array([[delxx,delyx],[delxy,delyy]])
    return(h)

args = (1,100)
x0 = [2,2]
```

```python
x = np.linspace(-2,2,100)
y = np.linspace(-2,2,100)
xx,yy = np.meshgrid(x,y)
zz = f([xx,yy],args)
fig = plt.figure(figsize=(6,4),dpi=150)
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xx,yy,zz,cmap=cm.coolwarm,linewidth=0, antialiased=False)
plt.tight_layout()
```

```python
result = minimize(fun = f,
                  x0=x0, 
                  args = (args,),
                  method="Nelder-Mead")
print(result)
```

```python
print(sp.optimize.OptimizeResult)
```

```python
result = minimize(fun = f,
                  x0=x0, 
                  args = (args,),
                  method="Powell")
print(result)
```

```python
result = minimize(fun = f,
                  x0=x0, 
                  args = (args,),
                  jac= grad,
                  method="CG")
print(result)
```

```python
result = minimize(fun = f,
                  x0=x0, 
                  args = (args,),
                  jac= grad,
                  bounds = ((2,None),(2,np.inf)),
                  method="L-BFGS-B")
print(result)
```

```python
def cons(x):
    return -(x + 1)
cons = (
    {'type': 'ineq', 'fun': cons}
)
```

```python
cons = ({"type":"eq","fun": lambda x: x[0]+x[1]-1})
result = minimize(fun = f,
                  x0=x0, 
                  args = (args,),
                  jac= grad,
                  bounds = ((0,1),) *2,
                  constraints= cons,
                  method="SLSQP")
print(result)
```
